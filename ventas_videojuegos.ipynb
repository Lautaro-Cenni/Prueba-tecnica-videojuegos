### Ventas por editorial

```python
ventas_por_editorial_ultimos_años = ventas_videojuegos[ventas_videojuegos.Año > 2012]
ventas_por_editorial_ultimos_años = ventas_por_editorial_ultimos_años[['Editorial','Ventas NA', 'Ventas EU', 'Ventas JP', 'Ventas Otros','Ventas Global']].groupby(by=['Editorial'], as_index=False).sum()
ventas_por_editorial_ultimos_años = ventas_por_editorial_ultimos_años.sort_values(by='Ventas Global', ascending=False)[:10]
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python
figure, axes = plt.subplots(4, 1, figsize=(15,18))
for i in range (1,5):
    x = ventas_por_editorial_ultimos_años.Editorial
    y = ventas_por_editorial_ultimos_años[ventas_por_editorial_ultimos_años.columns[i]]
    ax = sns.barplot(ax=axes[i-1],x = x, y = y)
    ax.set(xlabel='')
    wrap_labels(ax, 15)
    ax.figure
```

# Conclusión:

* hubo un decremento en las ventas en los videojuegos en los ultimos años, eso no implica que el negocio de los videojuegos haya emperoado necesariamente ya que este sufrio un cambio en el tipo de mercado que predomina (microtransacciones, juegos free to play, juego de movil con propagandas, etc)
* En Japon, uno de los principales paises consumidores, ocupando un poco mas del 14% de las compras totales en todo el mundo, en su mayoria solo consumen juegos provenientes de editoriales de Japon (Nintendo, Square Enix, Bandai Namco) y los generos mas comunes en japon son "role playing", "accion" y "deportes"
* En todos los paises excepto Japon los generos mas consumidos son "accion", "deportes", y "shooter"
* Las consolas van cambiando con el tiempo y es importante actualizarse, consolas que vendieron mucho como la PS2 pierden por completo relevancia al aparecer nuevas consolas.

## Voy a cambiar un poco los datos para que me sean mas comodos de operar en power bi 

### voy a separar el dataset en 5, donde cada dataset tendra las ventas en un pais en específico y uno que tendra las ventas globales

```python
ventas_videojuegos_NA = ventas_videojuegos.drop(columns= ventas_videojuegos.iloc[:,6:10], axis=1)
ventas_videojuegos_EU = ventas_videojuegos.drop(columns= ventas_videojuegos.iloc[:,[5,7,8,9]], axis=1)
ventas_videojuegos_JP = ventas_videojuegos.drop(columns= ventas_videojuegos.iloc[:,[5,6,8,9]], axis=1)
ventas_videojuegos_Otros = ventas_videojuegos.drop(columns= ventas_videojuegos.iloc[:,[5,6,7,9]], axis=1)
ventas_videojuegos_Global = ventas_videojuegos.drop(columns= ventas_videojuegos.iloc[:,5:9], axis=1)
datasets = [ventas_videojuegos_NA, ventas_videojuegos_EU, ventas_videojuegos_JP, ventas_videojuegos_Otros, ventas_videojuegos_Global]
```

```python
for index, dataset in enumerate(datasets):
    region = dataset.columns[-1].split()[1]
    dataset.rename(columns={dataset.columns[-1] : 'Ventas'}, inplace=True)
    dataset['Región'] = region
    if index == 0:
        ventas_videojuegos_final = dataset
    else:
        ventas_videojuegos_final = pd.concat([ventas_videojuegos_final,datasets[index]])
```
